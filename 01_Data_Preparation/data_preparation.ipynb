Change timestemp variables to datatime type

```python
charging_data["connectionTime"] = pd.to_datetime(charging_data["connectionTime"])
charging_data["disconnectTime"] = pd.to_datetime(charging_data["disconnectTime"])
charging_data["doneChargingTime"] = pd.to_datetime(charging_data["doneChargingTime"])
```

```python
weather_data["timestamp"] = pd.to_datetime(weather_data["timestamp"])
```

Check and remove duplicates

```python
print("Charging Data number of rows before removal: ", len(charging_data))
print("Number of duplicate rows in charing data: ", charging_data.duplicated().sum())
print("Number of duplicate rows in weather data: ", weather_data.duplicated().sum())
charging_data = charging_data.drop_duplicates()
weather_data = weather_data.drop_duplicates()
```

Check uniqueness of certain variables that need to be unique

```python
print(f"ID is unique: {charging_data['id'].is_unique}")
print(f"Unnamed: 0 is unique: {charging_data['Unnamed: 0'].is_unique}")
```

```python
doubles = charging_data["id"][charging_data["id"].duplicated()]
print("Double entries: ", doubles.tolist())
```
